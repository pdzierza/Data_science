# Analiza danych o postępach uczniów szkoły średniej

Poniższa analiza opiera się o dane zebrane w portugalskiej szkole im. Gabriela Pereiry w mieście Évora. Dane zebrane zostały w celu określenia czynników wpływających na wysokość końcowej oceny z matematyki uzyskanej przez ucznia. Do dyspozycji mamy następujące zmienne:

1) school - szkoła - nazwa szkoły, z której pochodzą dane (analizie poddana zostanie tylko szkoła GP)

2) sex - płeć ucznia - przez "F" zakodowano kobiety, a przez "M" mężczyzn

3) age - wiek ucznia - w latach

4) address - miejsce zamieszkania - przez "U" zakodowano ucznia mieszkającego w mieście, przez "R" ucznia ze wsi

5) famsize - liczba osób w rodzinie - "LE3" oznacza, że rodzina ucznia liczy 3 osoby lub mniej, "GT3" oznacza przynajmniej 3 osoby

6) Pstatus - status rodziców ucznia - "T" oznacza, że rodzice żyją razem, "A" oznacza, że żyją osobno

7) Medu - edukacja matki ucznia - kodowanie: "0" - brak jakiegokolwiek wykształcenia, "1" - skończone pierwsze 4 klasy, "2" – skończone 5 do 9 klas, "3" – wykształcenie średnie, "4" – wykształcenie wyższe

8) Fedu - edukacja ojca ucznia - kodowanie: "0" - brak jakiegokolwiek wykształcenia, "1" - skończone pierwsze 4 klasy, "2" – skończone 5 do 9 klas, "3" – wykształcenie średnie, "4" – wykształcenie wyższe

9) Mjob - praca matki - możliwe wartości: "teacher" - nauczyciel, "health" - praca w służbie zdrowia, "services" - służby publiczne, "at_home" - w domu, "other" - inny zawód

10) Fjob - praca ojca - możliwe wartości: "teacher" - nauczyciel, "health" - praca w służbie zdrowia, "services" - służby publiczne, "at_home" - w domu, "other" - inny zawód

11) reason - główny powód wyboru szkoły przez ucznia - możliwe wartości: "home" - blisko domu, "reputation" - reputacja szkoły, "course" - oferta edukacyjna szkoły, "other" - inny powód

12) guardian - opiekun prawny ucznia - możliwe wartości: "mother" - matka, "father" - ojciec, "other" - inna osoba

13) traveltime - czas podróży z domu do szkoły - kodowanie:  "1" - do 15 minut, "2" - 15 do 30 minut, "3" - 30 do 60 minut, "4" - powyżej 1 godziny

14) studytime - czas spędzony na nauce tygodniowo - kodowanie: "1" - poniżej 2 godzin, "2" - 2 do 5 godzin, "3" - 5 do 10 godzin, "4" - powyżej 10 godzin

15) failures - liczba oblanych przedmiotów przez ucznia do momentu badania (nie zostało sprecyzowane czy chodzi tylko o matematykę) - kodowanie: "1" - 1 powtórka, "2" - 2 powtórki, "3" - 3 powtórki, "4" - uczeń oblał więcej niż 3 razy

16) schoolsup - informacja o korzystaniu z dodatkowej pomocy edukacyjnej - kodowanie: "yes" - tak, "no" - nie

17) famsup - informacja o dodatkowym wsparciu materialnym rodziny ucznia - kodowanie: "yes" - tak, "no" - nie

18) paid - informacja o korzystaniu z dodatkowych płatnych lekcji matematyki - kodowanie: "yes" - tak, "no" - nie

19) activities - informacja o uczęszczaniu na dodatkowe zajęcia pozalekcyjne - kodowanie: "yes" - tak, "no" - nie

20) nursery - informacja o tym, czy uczeń chodził do żłobka - kodowanie: "yes" - tak, "no" - nie

21) higher - informacja czy uczeń planuje iść na studia - kodowanie: "yes" - tak, "no" - nie

22) internet - informacja o dostępie do Internetu przez ucznia w domu - kodowanie: "yes" - tak, "no" - nie

23) romantic - informacja, czy uczeń był w czasie badania w związku (czy ma partnera) - kodowanie: "yes" - tak, "no" - nie

24) famrel - jakość stosunków rodzinnych - kodowanie: "5" - doskonałe, "4"- dobre, "3" - średnie, "2" - złe, "1"- bardzo złe

25) freetime - ilość wolnego czasu ucznia - kodowanie: "5" - bardzo dużo, "4" - dużo, "3" - średnio, "2" - mało, "1"- bardzo mało

26) goout - częstość spotykania się ze znajomymi - kodowanie: "5" - bardzo często, "4" - często, "3" - czasami, "2" - rzadko, "1"- bardzo rzadko

27) Dalc - spożycie alkoholu przez ucznia w tygodniu (poniedziałek - piątek) - kodowanie: "5" - bardzo duże, "4" - duże, "3" - średnie, "2" - małe, "1"- bardzo małe

28) Walc - spożycie alkoholu przez ucznia podczas weekendu - kodowanie: "5" - bardzo duże, "4" - duże, "3" - średnie, "2" - małe, "1"- bardzo małe

29) health - ogólna samoocena zdrowia ucznia w momencie badań - "5" - doskonałe, "4" - dobre, "3" - średnie, "2" - złe, "1" - bardzo złe

30) absences - liczba nieobecności ucznia w roku szkolnym

31) G1 - liczba punktów (ocena) za pierwszy semestr - wartość między 0 a 20

32) G2 - liczba punktów (ocena) za drugi semestr - wartość między 0 a 20

33) G3 - końcowa liczba punktów (ocena) - wartość między 0 a 20



W tym raporcie analizie zostanie wykonany przegląd wszystkich powyższych pomiarów. Głównym zainteresowaniem będą tutaj informacje dotyczące słabszych uczniów. Przyjęte zostało, że słabszy uczeń to taki, którego ocena końcowa, G3, była mniejsza od 10.

Następnie podjęta zostanie próba porównania ze sobą średnich wartości ocen G1, G2 i G3. Celem będzie sprawdzenie, czy różnice między nimi są statystycznie istotne.

Kolejnym krokiem będzie budowa modelu, którego zadaniem będzie przewidzenie wartości G3 ucznia, przy użyciu innych dostępnych informacji o nim, lecz bez użycia G1 i G2. Jest to główny powód, dla którego zebrano użyte tutaj dane.

Na potrzeby wizualizacji i dalszego modelowania stworzono drugi zbiór danych, w którym skorzystano z dummy variables jako kodowanie binarnych zmiennych. 

```python
dummy_cols = [cname for cname in dane.columns if
                    dane[cname].nunique() ==2  and 
                    dane[cname].dtype == "object"]

student=pd.get_dummies(dane,columns=dummy_cols,drop_first=True)
```

```python
student.loc[student.guardian=='other','guardian']=0
student.loc[student.guardian!=0,'guardian']=1
student=student.rename(columns={"guardian":"guardianParent"})
student = student.astype({"guardianParent": int})
```

## Próba porównania średnich wyników G1, G2 i G3 osób, które nie zrezygnowały z próby zaliczenia

Porównanie ze sobą średnich wyników zaliczeń może pomóc w określeniu, czy któryś z etapów (pierwszy semestr, drugi semestr, egzaminy końcowe) jest zauważalnie trudniejszy od innych. Tutaj dane zostaną ograniczone do osób, które uzyskały przynajmniej 1 punkt z G3. Takie ograniczenie nie dość, że potencjalnie da bardziej interpretowalne wyniki, to również pozwala pozbyć się problematycznego punktu 0 w G3, który koncentruje w sobie 10% całej populacji badanych uczniów.

Jedną z metod porównana ze sobą śrenich kilku różnych pomiarów jednego zjawiska jest Analiza Wariancji Powtórzonych Pomiarów (rANOVA). Metoda ta ma takie same założenia jak zwykła ANOVA. Pierwszym krokiem analizy będzie sprawdzenie normalności G1, G2 i G3 testem Shapiro-Wilka na standardowym poziomie istotności równym 0.05.

```python
BezZer=student[student.G3!=0]
```

```python
shapiroG1 = stats.shapiro(BezZer.G1)
shapiroG2 = stats.shapiro(BezZer.G2)
shapiroG3 = stats.shapiro(BezZer.G3)
print("p-wartość testu Shapiro-Wilka dla G1: ", shapiroG1[1])
print("p-wartość testu Shapiro-Wilka dla G2: ", shapiroG2[1])
print("p-wartość testu Shapiro-Wilka dla G3: ", shapiroG3[1])
```

Wszystkie trzy testy orzekły, że na poziomie istotności 0.05, należy odrzucić hipotezę zerową, że wyniki G1, G2 lub G3 pochodzą z rozkładów normalnych. W takiej sytuacji pozostaje spróbować wykonać transformację danych. Pierwszym podejściem będzie nałożenie logarytmu naturalnego na każdą wartość G1, G2 i G3.

```python
shapiroLogG1 = stats.shapiro(np.log(BezZer.G1))
shapiroLogG2 = stats.shapiro(np.log(BezZer.G2))
shapiroLogG3 = stats.shapiro(np.log(BezZer.G3))
print("p-wartość testu Shapiro-Wilka dla log(G1): ", shapiroLogG1[1])
print("p-wartość testu Shapiro-Wilka dla log(G2): ", shapiroLogG2[1])
print("p-wartość testu Shapiro-Wilka dla log(G3): ", shapiroLogG3[1])
```

Ponownie, test Shapiro-Wilka, na poziomie istotności równym 0.05, orzeka, że dane po transformacji logarytmicznej wciąż nie należą do rozkładów normalnych. Drugim potencjalnym rozwiązaniem jest wzięcie pierwiastka z wyników.

```python
shapiroSqrtG1 = stats.shapiro(np.sqrt(BezZer.G1))
shapiroSqrtG2 = stats.shapiro(np.sqrt(BezZer.G2))
shapiroSqrtG3 = stats.shapiro(np.sqrt(BezZer.G3))
print("p-wartość testu Shapiro-Wilka dla sqrt(G1): ", shapiroSqrtG1[1])
print("p-wartość testu Shapiro-Wilka dla sqrt(G2): ", shapiroSqrtG2[1])
print("p-wartość testu Shapiro-Wilka dla sqrt(G3): ", shapiroSqrtG3[1])
```

Ponownie, test Shapiro-Wilka, na poziomie istotności równym 0.05, orzeka, że dane po nałożeniu pierwiastka wciąż nie należą do rozkładów normalnych. 

Ostatnim podejściem będzie próba dopasowania transformacji Boxa-Coxa przy parametrze $\lambda$ różnym od zera. Do sprawdzenia, jaka wartość lambda jest odpowiednia, przydadzą się histogramy oraz wykresy pudełkowe.

```python
fig = go.Figure(data=[go.Histogram(x=BezZer.G1)])



fig.update_layout(
    title_text="Rozkład wyników G1 uczniów, którzy nie zrezygnowali",
     xaxis_title_text='Wynik',
     yaxis_title_text='Liczba osób')


fig.show()
```

```python
fig = go.Figure(data=[go.Histogram(x=BezZer.G2)])


fig.update_layout(
    title_text="Rozkład wyników G2 uczniów, którzy nie zrezygnowali",
     xaxis_title_text='Wynik',
     yaxis_title_text='Liczba osób')


fig.show()
```

```python
fig = go.Figure(data=[go.Histogram(x=BezZer.G3)])


fig.update_layout(
    title_text="Rozkład wyników G3 uczniów, którzy nie zrezygnowali",
     xaxis_title_text='Wynik',
     yaxis_title_text='Liczba osób')


fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Box(y=BezZer.G1, name="G1"))
fig.add_trace(go.Box(y=BezZer.G2, name="G2"))
fig.add_trace(go.Box(y=BezZer.G3, name="G3"))

fig.update_layout(
    title_text="Boxploty dla uczniów, którzy nie zrezygnowali",
     xaxis_title_text='Zaliczenie')

fig.show()
```

Z wykresów można odczytać, że dane są prawostronnie asymetryczne (przeważają mniejsze obserwacje). Potrzebna będzie zatem transformacja wklęsła, która charakteryzuje się wartością $\lambda$ mniejszą od 1.

```python
lam=0.7
shapiroBCG1 = stats.shapiro(((BezZer.G1)**lam-1)/lam)
shapiroBCG2 = stats.shapiro(((BezZer.G2)**lam-1)/lam)
shapiroBCG3 = stats.shapiro(((BezZer.G3)**lam-1)/lam)
print('Dla lambdy równej ', lam,":")
print("p-wartość testu Shapiro-Wilka po transformacji Boxa-Coxa dla G1: ", shapiroBCG1[1])
print("p-wartość testu Shapiro-Wilka po transformacji Boxa-Coxa dla G2: ", shapiroBCG2[1])
print("p-wartość testu Shapiro-Wilka po transformacji Boxa-Coxa dla G3: ", shapiroBCG3[1])
```

Próba dobrania wartości parametru $\lambda$ tak, by uzyskać dane pochodzące z rozkładu normalnego, zakończyła się niepowodzeniem. Rezultat ten był do przewidzenia - transformacja Boxa-Coxa rzadko sprawdza się w przypadku danych dyskretnych.

Należy wyciągnąć wniosek, że rANOVA nie pasuje do danych. Pozostaje zatem sięgnąć do innych, mniej standardowych metod. Jedną z nich jest test Friedmana. Nie wymaga on założeń metody ANOVA, lecz w przypadku małej liczby grup (poniżej 5) należy mieć świadomość, że oszacowanie p-wartości testu może być niedokładne.

```python
print("Wynik testu Friedmana:")
stats.friedmanchisquare(BezZer.G1,BezZer.G2,BezZer.G3)
```

Na podstawie testu Friedmana na poziomie istotności 0.05 należy odrzucić hipotezę zerową o równości wszystkich trzech średnich wyników G1, G2 i G3. W celu wykrycia, które grupy są od siebie różne, zastosowana zostanie metoda Dunna analizy post hoc.

```python
print('Wynik procedury post hoc Dunna:')
scikit_posthocs.posthoc_dunn([BezZer.G1,BezZer.G2,BezZer.G3], p_adjust = "holm")
```

Wyniki przedstawione w tabeli powyżej opisują p-wartości otrzymane w wyniku porównania metodą Dunna parami każdej grupy. Wyniki na przekątnej pokazują porównania każdej grupy z samą sobą i nie wnoszą żadnej informacji. Wykorzystana tu analiza post hoc bazuje na porównaniach wielokrotnych, w celu korekty p-wartości zastosowano poprawkę Holma zamiast standardowej poprawki Bonferoniego. Mimo to wyniki przedstawiają sytuację, w której wszystkie porównania wskazują brak różnic między grupami. Rezultat ten jest sprzeczny z wynikami otrzymanymi z testu Friedmana. Mogą być dwie główne przyczyny tej rozbieżności. Pierwszą jest wspomniane już wcześniej niedoszacowanie p-wartości testu Friedmana, gdy porównuje się ze sobą małą liczbę grup. Drugą przyczyną może być niska moc testu w analizie Dunna, która nie pozwala na odrzucenie żadnych hipotez zerowych o braku istotnych różnic między grupami. Należy zatem wnioskować, że równość G1, G2 i G3 jest możliwa, choć jest niepewna.

## Predykcja G3 bez użycia G1 i G2

W tej sekcji przedstawione zostanie jedno z możliwych podejść przewidzenia wyniku G3 ucznia, korzystając tylko z dostępnych danych z wyłączeniem wyników G1 i G2. Wykorzystany zostanie model Extreme Boosting (XGBoost), który wykorzystuje drzewa losowe. Z tego powodu skalowanie zmiennych jest niepotrzebne. Do wyznaczenia hiperparametrów modelu wykorzystane zostanie połączenie kroswalidacji z 5 foldami i losowe przeszukiwanie siatki hiperparametrów. Model zostanie wyuczony przy użyciu 80% zbioru danych wybranego przez losowanie. Pozostałe 20% posłuży jako zbiór testowy. Metryką błędu oceniającą jakość predykcji będzie Root Mean Squared Error (RMSE).

Zmienne zależne, które zostaną użyte w modelu to: liczba porażek ('failures'), indykator płci męskiej ('sex_M'), stopień wykształcenia matki ('Medu'), stopień częstości wychodzenia ucznia ze znajomymi ('goout'), indykator planowania pójścia na studia ('higher_yes'), stopień czasu spędzanego na nauce('studytime'), stopień spożycia alkoholu w weekendy ('Walc'). 

```python
X_train, X_test, y_train, y_test = train_test_split(student[['failures','sex_M','Medu','goout','higher_yes','studytime','Walc']], student.G3, 
                                                                train_size=0.8, test_size=0.2,
                                                                random_state=0)

paramsXG = {
        'min_child_weight': [1, 5, 10],
        'gamma': [0,0.5,1, 5],
        'subsample': [0.7, 1.0],
        'colsample_bytree': [0.7, 1.0],
        'max_depth': [4, 5,6,7],
        'reg_lambda' : [1,2,3],
        'eta' : [0.3,0.6]
        }

XGbReg = XGBRegressor(n_estimators=1000, learning_rate=0.01, random_state=1001) 
```

Model XGBoost w wybranymi hiperparametrami osiągnął na zbiorze testowym wartość RMSE równą 4.58592881499018. Wynik ten może wydawać się mało zadowalający. Można jednak spojrzeć, dla jakich przypadków metoda dała najgorsze wyniki.

## Wyznaczenie grupy uczniów słabszych przy pomocy analizy skupień

W tej części wykorzystana zostanie metoda k średnich do stworzenia pogrupowania uczniów względem wyników. Pod uwagę zostaną wzięte wyniki G1 i G3. Przy wyborze liczby klastrów jako funkcja straty zostanie użyta suma kwadratów odległości punktów do przypisanych im finalnych centroidów (WCSS).

```python

wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(dane[['G1','G3']])
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Liczba klastrów')
plt.ylabel('wcss')
plt.show()
```

Z powyższego wykresu można odczytać, że optymalną liczbą klastrów jest liczba 4.

```python
kmeansmodel = KMeans(n_clusters= 4, init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(dane[['G1','G3']])
```

```python
dane.insert(len(dane.columns),'klaster',y_kmeans,True)
```

```python
plt.scatter(dane[dane['klaster']==0]['G1'], dane[dane['klaster']==0]['G3'], s = 100, c = 'red', label = 'Klaster 1')
plt.scatter(dane[dane['klaster']==1]['G1'], dane[dane['klaster']==1]['G3'], s = 100, c = 'green', label = 'Klaster 2')
plt.scatter(dane[dane['klaster']==2]['G1'], dane[dane['klaster']==2]['G3'], s = 100, c = 'blue', label = 'Klaster 3')
plt.scatter(dane[dane['klaster']==3]['G1'], dane[dane['klaster']==3]['G3'], s = 100, c = 'purple', label = 'Klaster 4')
plt.scatter(kmeansmodel.cluster_centers_[:, 0], kmeansmodel.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroidy')
plt.title('Podział uczniów na grupy ze względu na wyniki G1 i G3')
plt.xlabel('Wynik G1')
plt.ylabel('Wynik G3')
plt.legend()
plt.show()
```

Powyższy wykres obrazuje 4 grupy:
    
- niebieska to osoby, które zrezygnowały,

- czerwona to osoby słabsze,

- zielona to osoby średnie,

- fioletowa to osoby zdolne.

Z tego podziału wyłamuje się jedna osoba, która mimo słabych, lecz niezerowych wyników, wciąż jest w grupie niebieskiej. Można by było zatem uznać, że osoba słabsza to uczeń należący do dwóch pierwszych klastrów albo tylko do klastra drugiego.
